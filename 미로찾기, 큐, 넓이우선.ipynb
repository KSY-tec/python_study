```python
#큐
class QueueADT():
  def __init__(self):
      pass
  def Queue(self):
    self.items=[]
    return self.items
  def isEmpty(self):
    return len(self.items)==0
  def enqueue(self,item):
    self.items.append(item)
  def dequeue(self):
    if not self.isEmpty():
      return self.items.pop(0)
  def peek(self):
    if not self.isEmpty():
      return self.items[-1]
  def size(self):
    return len(self.items)
  def clear(self):
    self.items=[]
```
