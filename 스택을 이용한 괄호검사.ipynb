```python
class StackADT:
  def __init__(self):
    pass
  def Stack(self):
    self.stack=[]

  def isEmpty(self):
    if len(self.stack)==0:
      return True
    else:
      return False
  def push(self, e):
    self.stack.append(e)
  def pop(self):
    temp=self.stack[-1]
    self.stack.remove(self.stack[-1])
    return temp
  def peek(self):
    return self.stack[-1]
  def size(self):
    len(self.stack)
  def clear(self):
    self.stack.clear()
```
