```python
class StackADT:
  def __init__(self):
    pass
  def Stack(self):
    self.stack=[]

  def isEmpty(self):
    if len(self.stack)==0:
      return True
    else:
      return False
  def push(self, e):
    self.stack.append(e)
  def pop(self):
    temp=self.stack[-1]
    self.stack.remove(self.stack[-1])
    return temp
  def peek(self):
    return self.stack[-1]
  def size(self):
    len(self.stack)
  def clear(self):
    self.stack.clear()
```

```python
map=[
     ['1','1','1','1','1','1'],
     ['e','0','0','0','0','1'],
     ['1','0','1','0','1','1'],
     ['1','1','1','0','0','x'],
     ['1','1','1','0','1','1'],
     ['1','1','1','1','1','1'],
     ]
MAZE_SIZE=6
```

```python
#경로 검색 함수
def wheretogo(x,y):
  temp=[]
  map[x][y]='z'
  print("현재 : map["+str(x)+","+str(y)+"]")
  if y+1<=len(map[0]):
    #print("동 : map["+str(x)+","+str(y+1)+"]")
    #print("벽 여부 : "+str(map[x][y+1]))
    if map[x][y+1]=='0':
      temp+=[(x, y+1)]
    elif map[x][y+1]=='x':
      print("출구는 "+str(x)+', '+str(y+1))
      #return ["exit"]
  else:
    print("동 : map size 범위를 벗어남")
  if y-1>=0:
    #print("서 : map["+str(x)+","+str(y-1)+"]")
    #print("벽 여부 : "+str(map[x][y-1]))
    if map[x][y-1]=='0':
      temp+=[(x, y-1)]
    elif map[x][y-1]=='x':
      print("출구는 "+x+', '+(y-1))
      #return ["exit"]
  else:
    print("서 : map size 범위를 벗어남")
  if x+1<=len(map):
    #print("남 : map["+str(x+1)+","+str(y)+"]")
    #print("벽 여부 : "+str(map[x+1][y]))
    if map[x+1][y]=='0':
      temp+=[(x+1, y)]
    elif map[x+1][y]=='x':
      print("출구는 "+str(x+1)+", "+str(y))
      #return "exit"
  else:
    print("남: map size 범위를 벗어남")
  if x-1>=0:
    #print("북 : map["+str(x-1)+","+str(y)+"]")
    #print("벽 여부 : "+str(map[x-1][y]))
    if map[x-1][y]=='0':
      temp+=[(x-1, y)]
    elif map[x-1][y]=='x':
      print("출구는"+str(x-1)+', '+str(y))
      #return "exit"
  else:
    print("북 : map size 범위를 벗어남")
  
  return temp
```

```python
#시작점 검색 함수
def startat(geog):
  for i in range(len(geog)):
    for j in range(len(geog)):
      if geog[i][j]=="e":
        return i, j
```

```python
#스택 공간 생성
stack=StackADT()
stack.Stack()
```

```python
#시작점 탐색
start=startat(map)
start
```

```python
for i in (wheretogo(start[0],start[1])):
  stack.push(i)
```

```python
#깊이탐색 시행
while True:
  try:
    ind=stack.pop()
    for i in (wheretogo(ind[0],ind[1])):
      stack.push(i)
  except:
    print("종료")
    break
```
