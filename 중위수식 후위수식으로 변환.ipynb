```python
class StackADT:
  def __init__(self):
    pass
  def Stack(self):
    self.stack=[]

  def isEmpty(self):
    if len(self.stack)==0:
      return True
    else:
      return False
  def push(self, e):
    self.stack.append(e)
  def pop(self):
    temp=self.stack[-1]
    self.stack.remove(self.stack[-1])
    return temp
  def peek(self):
    return self.stack[-1]
  def size(self):
    len(self.stack)
  def clear(self):
    self.stack.clear()
```

```python
test=[
"A+B*C",
"A*B+C",
"(A+B)*C"
]
```

```python
def transform(calto):
  temp=""
  stck=StackADT()
  stck.Stack()
  for i in calto:

    if i in "(":
      continue
    if i in ")":
      temp+=stck.pop()
      continue
      
    if i in "*+-/":
      if len(stck.stack)>0:
        if i in "*/" and stck.stack[-1] not in "*/":
          stck.push(i)
        else:
          temp+=stck.pop()
          stck.push(i)
      else:
        stck.push(i)
    else:
      temp+=i


  if len(stck.stack)>0:
    for i in range(len(stck.stack)):
      temp+=stck.pop()

  return temp
```

```python
transform(test[0])
```
